```python
import cv2
import numpy as np
import mediapipe as mp
import csv
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
%matplotlib inline
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn import svm
```

```python
# Initializing mediapipe pose class.
mp_pose = mp.solutions.pose

# Initializing mediapipe drawing class, useful for annotation.
mp_drawing = mp.solutions.drawing_utils 
```

```python
landmarks2 = ['class']
for val in range(11, 33):
    landmarks2 += ['x{}'.format(val), 'y{}'.format(val), 'z{}'.format(val), 'v{}'.format(val)]

with open('pose.csv', mode='w', newline='') as f:
    csv_writer = csv.writer(f, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)
    csv_writer.writerow(landmarks2)
```

```python
directory_images = 'D:\\semester 7\\Image processing\\project\\image datasetss\\1- IMAGES'
folders = []
for filename in os.listdir(directory_images):
    f = os.path.join(directory_images, filename)
    folders.append(f)
    
```

```python
pose = mp_pose.Pose(static_image_mode=True, min_detection_confidence=0.3, model_complexity=0)

for folder in folders:
    for imagename in os.listdir(folder):
        # print(folder[65:], imagename)
        test = cv2.imread(os.path.join(folder, imagename))
        output_image = test.copy()
        imageRGB = cv2.cvtColor(test, cv2.COLOR_BGR2RGB)


        results = pose.process(imageRGB)
        # landmarks = []
        print(imagename)

        if results.pose_landmarks:
            resultpose = results.pose_landmarks.landmark
            landmarks = list(
                np.array([[landmark.x, landmark.y, landmark.z, landmark.visibility]
                          for landmark in resultpose]).flatten())

            class_name = folder[65:]
            landmarks = landmarks[44:]

            landmarks.insert(0, class_name)
            #print(landmarks)
            with open('pose.csv', mode='a', newline='') as f:
                csv_writer = csv.writer(f, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)
                csv_writer.writerow(landmarks)
    print("Done", folder)
```

```python
directory_videos = 'D:\\semester 7\\Image processing\\project\\image datasetss\\2- VIDEOS'

folders = []
for filename in os.listdir(directory_videos):
    f = os.path.join(directory_videos, filename)
    folders.append(f)

for folder in folders:
    for videoname in os.listdir(folder):
        cap = cv2.VideoCapture(os.path.join(folder, videoname))
        with mp_pose.Pose(static_image_mode=False, min_detection_confidence=0.5, model_complexity=1) as holistic:
            sec = 0
            frameRate = 0.2
            while cap.isOpened():
                sec = sec + frameRate
                sec = round(sec, 2)
                cap.set(cv2.CAP_PROP_POS_MSEC, sec * 1000)
                ret, frame = cap.read()
                if ret == True:
                    results = holistic.process(frame)
                    try:
                        # Extract Pose landmarks
                        pose = results.pose_landmarks.landmark
                        pose_row = list(np.array(
                            [[landmark.x, landmark.y, landmark.z, landmark.visibility] for landmark in pose]).flatten())

                        pose_row = pose_row[44:]
                        print(videoname)

                        class_name = folder[65:]

                        # Append class name
                        pose_row.insert(0, class_name)

                        # Export to CSV
                        with open('pose.csv', mode='a', newline='') as f:
                            csv_writer = csv.writer(f, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)
                            csv_writer.writerow(pose_row)
                    except:
                        pass

                    if cv2.waitKey(10) & 0xFF == 27:
                        break
                else:
                    break
        print("Done", folder)
```

```python
def rescale_frame(frame, percent=75):
    width = int(frame.shape[1] * percent/ 100)
    height = int(frame.shape[0] * percent/ 100)
    dim = (width, height)
    return cv2.resize(frame, dim, interpolation =cv2.INTER_AREA)
```

```python
mp_face_mesh = mp.solutions.face_mesh
face_mesh = mp_face_mesh.FaceMesh()
```

```python
landmarks2 = ['class']
for val in range(0, 468):
    landmarks2 += ['x{}'.format(val), 'y{}'.format(val), 'z{}'.format(val)]

with open('face.csv', mode='w', newline='') as f:
    csv_writer = csv.writer(f, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)
    csv_writer.writerow(landmarks2)
```

```python
directory_images = 'D:\\semester 7\\Image processing\\project\\filtered'

folders = []
for filename in os.listdir(directory_images):
    f = os.path.join(directory_images, filename)
    folders.append(f)
    
```

```python
for folder in folders:
    for imagename in os.listdir(folder):
            test = cv2.imread(os.path.join(folder, imagename))
            output_image = test.copy()
            imageRGB = cv2.cvtColor(test, cv2.COLOR_BGR2RGB)
            
            result = face_mesh.process(imageRGB)
            if result.multi_face_landmarks:
                for facial_landmarks in result.multi_face_landmarks:
                    final = []
                    for i in range(0,468):
                        pt1 = facial_landmarks.landmark[i]
                        x = pt1.x 
                        y= pt1.y 
                        z= pt1.z 
                        final.append(x)
                        final.append(y)
                        final.append(z)

                        #print(imagename, x,y,z)
                        class_name = folder[48:]
                    final.insert(0, class_name)
                    #print(final)
                
                    with open('face.csv', mode='a', newline='') as f:
                        csv_writer = csv.writer(f, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)
                        csv_writer.writerow(final)
    print("Done", folder)
```
